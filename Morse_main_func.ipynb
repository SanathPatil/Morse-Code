```python
from decoder import Decoder
from character import CharacterAnalyser
from word import WordAnalyser
from sentence import SentenceAnalyser

decoder_obj=Decoder()
charCount_obj=CharacterAnalyser()
wordCount_obj=WordAnalyser()
sentCount_obj=SentenceAnalyser()

#method to take valid input
def valid_seq():
    total_seq=''
    while(True):
        code=('1','0','*','***',',')
        
        morse=input('Please enter a valid morse code (0,1,* or ***-as delimeter for letter and word) to be decoded or press q+enter to quit:')
        
        if (len(morse)==0):
            print('You have to enter atleast one valid sequence as the sequence cannot be empty')
        elif (morse=='q'):
            print('Exiting the user from entering the code!')
            break
        elif all((x in code) for x in morse):
            
            if(morse.endswith('001100') or morse.endswith('110011')or morse.endswith('010101')):
                total_seq+=morse
                total_seq+=','
            else:
                print('Invalid code!!! The sentence must end with a clause-(.,?)')
                continue
                
        else:
            print('Invalid code entered, only enter 0,1 ,* or *** to be accepted!!!')
    # To remove ',' from the end of the string attached earlier
    remove=total_seq.rfind(",")
    final_total_seq= total_seq[:remove]+''+total_seq[remove+1:]
    print('The final code entered is :{}'.format(final_total_seq))
    return final_total_seq


def main():
    code=valid_seq()
    #calling decode method to decode the morse code
    full_decoded_seq=decoder_obj.decode(code)
    print(full_decoded_seq)
    print('---------------------------------------------------------------------------------')
    
    #method to choose the requiered analyser
    while(True):
        
        print('Menu option for analysis:-')
        option=input('Type c-character/w-word/s-sentnece/type q to quit:')
        if (option == 'c'):
            charCount_obj.analyse_characters(full_decoded_seq)
            print(charCount_obj)
            
    
        elif (option=='w'):
            wordCount_obj.analyse_words(full_decoded_seq)
            print(wordCount_obj)
    
        elif(option=='s'):
            sentCount_obj.analyse_sentences(full_decoded_seq)
            print(sentCount_obj)           
            
        elif(option=='q'):
            print(' -----------You are exited-------')
            break
        else:
            print('Invalid!!! Please chose a valid choice')
            continue

if __name__ == "__main__":
    main()
    
```
